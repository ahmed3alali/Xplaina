```python
import pandas as pd

coursera = pd.read_csv("Coursera.csv")
udemy = pd.read_csv("Udemy.csv")

# Make sure both have the same columns
common_cols = ['title', 'skills', 'rating', 'level', 'duration', 'instructor','description','platform']
courses = pd.concat([coursera[common_cols], udemy[common_cols]], ignore_index=True)

courses.to_csv("courses.csv", index=False)

theCourses = pd.read_csv("courses.csv")
theCourses
```
